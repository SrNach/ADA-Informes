#Problemas NP-Hard

###¿Qué quiere decir NP-Hard?
Para contextualizar, existen distintos tipos de problemas, y formas de categorizarlos:  
- Los problemas más sencillos que pueden ser resueltos en tiempo polinomial pertenecen al conjunto **P**.  
- El conjunto **NP** son los problemas que no pueden ser resueltos, pero si comprobar si existe su solución en tiempo polinomial.  
- Los problemas **NP-Completo** son los problemas más difíciles dentro del conjunto **NP**, son aquellos que no pueden ser resueltos en tiempo polinomial.

Finalmente, los problemas **NP-Hard** son problemas de decisión que contiene los problemas H tales que todo problema L en NP puede ser transformado polinomialmente en H.

Los problemas **NP-Hard** son dificiles de resolver utilizando métodos tradicionales de cálculo. Son complicados de resolver implican una gran cantidad de posibles soluciones que deben ser comprobadas para encontrar la mejor solución.

###¿Qué quiere decir que $P \neq NP$

![](https://www.researchgate.net/publication/336890186/figure/fig13/AS:822100717744138@1573015493195/Diagram-of-intersection-among-classes-P-NP-NP-complete-and-NP-hard-problems.png)

Esta conjetura, si bien se cree que es verdadera, no ha sido comprobada.

Sabemos que **P** $⊆$ **NP**, y que todos los problemas **P** pueden ser resuletos en tiempo polinomial, pero esto no significa que todos los problemas **NP** cumplan lo mismo.

La cuestión dentro de esta conjetura es que el conjunto **NP** es amplio, y no todo problema $\in$ **NP** puede ser resuelto en tiempo polinomial. Para que **P** $=$ **NP**, todos y cada uno de los problemas dentro del conjunto **NP** debe poder resolverse en tiempo polinomial. Dentro de **NP**, también existen problemas **NP-Hard**, y mientras estos no puedan ser resueltos en tiempo polinomial, **NP** nunca va a ser igual a **P**. 

###¿Cómo se prueba que un problema es NP-Hard usando reducciones?
El método de reducción consiste en que, al tener un problema $A$, este pueda ser transformado a uno $B$, y que esta transformación tenga tiempo de ejecución polinomial.  

![](https://cdn.discordapp.com/attachments/1029906268028145688/1043688664665960558/unknown.png)

Para comprobar si un problema $B$ pertenece al conjunto **NP-Hard**, primero se debe escoger un problema $A$, el cuál estemos seguros que este sea **NP-Hard**.  
Si este problema $A$ puede ser transformado a $B$, entonces $B \in $ **NP-Hard**. 

![](https://cdn.discordapp.com/attachments/1029906268028145688/1043689611517501450/unknown.png)

Supongamos que $Α$ es **NP-hard**, esto significa que encontrar un algoritmo polinomial para $A$ refutaría la conjetura **P** $≠$ **NP**. Pero si también existiera un algoritmo polinomial para resolver $B$, esto implicaría en $A$ (porque $A$ se reduce en $B$) y eso también refuta la conjetura **P** $≠$ **NP**, en otras palabras $B$ también es **NP-Hard**.

#Problema Knapsack
**Entrada:** Conjunto de valores $v = [v_1, v_2, ..., v_n]$ y de pesos $p = [p_1, p_2, ..., p_n]$, representando $n$ objetos, y $W$ la capacidad máxima de la mochila.  
**Salida:** Conjunto de objetos que resultan en el valor máximo, con peso total $\leq W$.

Knapsack, o el problema de la mochila, consiste encontrar una combinación de objetos que resulte en el mayor valor posible, que no supere el peso límite de la mochila. 

![](https://miro.medium.com/max/684/0*3dS6Jw8NzzSD-mn8.jpg)

###Problema NP-Hard (Subset Sum)
El problema NP-Hard que utilizaremos es el problema de la Suma de Subconjuntos.  
**Entrada:** Conjunto $S$ compuesto por números enteros y $T$, el valor o la suma buscada.    
**Salida:** Subconjunto $S' \subseteq S$, donde la suma de todos sus elementos sea igual al valor $T$ buscado.

Si se entrega un conjunto $S$ de enteros y un valor $T$, se busca si existe un subconjunto $S' ⊆ S$ tal que la suma de todos sus elementos sea igual al valor buscado $T$.

###Reducción
Para el knapsack se requiere maximizar la ganancia obtenida con un límite de peso de la mochila y un valor establecido para cada objeto por lo que el problema depende de dos variables.

Se considera que para cada objeto, el valor y el peso son igual a cada número del conjunto de entrada de subset-sum.

Si se da el caso en el que se tiene un conjunto $S$ de números enteros, este mismo conjunto $S$ correponderá a una secuencia de $n$ objetos, donde cada elemento del conjunto tendrá un peso $p$ y un valor $v$ igual al valor de cada número de $S$, donde $T$ es el valor buscado en la suma de los elementos del subconjunto $S^{\prime}$ (subset-sum) que será igual a la capacidad máxima $W$ de la mochila.

Al resolver el problema de la mochila se buscará maximizar la ganancia obtenida con los objetos o datos disponibles del conjunto $n$. Si se encuentra el caso de que $p{i}=v{i}=S_{i}$, entonces se habrá encontrado el subconjunto $S^{\prime} \subseteq S$, el cual estará compuesto por la suma máxima $h \leq T$. Si $h = T$, entonces $S$ corresponde a una solución para el problema de la suma de subconjuntos.

**Pre-Proceso**  
Se duplica la lista de valores, pero esta será utilizada para los pesos de los elementos.

**Proceso**  
Se aplica knapsack, en cuál soluciona el problema de la mochila y se obtiene la configuración que resulta en la suma máxima de valores posibles sin sobrepasar el límite de la mochila.

**Post-Proceso**  
Si el resultado de knapsack es igual al valor requerido, este retornará la lista, si esto no es así, se dará aviso de que no existe una combinación que sea similar al valor que se está buscando.

#Algoritmos Sencillos

###Problema de la mochila
Se puede resolver utilizando un método voraz (Greedy).
1. Primero, se crea una lista para almacenar la densidad del valor del objeto ($valor/peso$), y a su vez, también se crea una lista de índices que referencie a cada objeto.
2. Se ordenan los índices en relación a la densidad $valor/peso$, de menor a mayor.
3. De esta manera, se podrán ir agregando uno a uno, los objetos que más valgan la pena, ya que mientras mayor densidad tenga el objeto, más valor y menos peso tendrá.
4. Mientras que no pase el límite de la mochila, se pueden ir agregando los pbjetos en orden de la lista ordenada.

El tiempo de ejecución de este algoritmo depende del algoritmo que se utilice para ordenar la lista de índices, ya los pasos a seguir del algoritmo son sencillos, lo más complejo del proceso es el ordenamiento de datos, por lo que la complejidad temporal sera igual al del algoritmo de ordenamiento utilizado.

###Problema del Clique
El algoritmo Bron-Kerbosch resuelve el problema del clique, y veremos la varación que aplica pivotes para una ejecución más ordenada:

Se elige un nodo como pivote. Un pivote $u$ se elige y se encuentra cualquier máximo clique que incluya $u$ o uno de los nodos no adyacentes, si esto no es así, el clique aumentaría al agregar $u$. Esto se hace para evitar hacer llamadas recursivas de más. 

1. Crea un arreglo de booleanos, donde su tamaño es igual a la cantidad de arcos, sirve para saber si los nodos han sido visitados o no.
2. Se realiza un primer ciclo iterativo, para comprobar si el arco actual ha sido visitado o no.
3. Si el arco actual no ha sido visitado aún, se entra a un segundo ciclo iterativo anidado, donde, si los arcos que se enlazan con el nodo no han sido visitados, se marcan como visitados y se marca el arco principal como visitado.

El tiempo de ejecución de este algoritmo es de $O(V + E)$, esto es debido a que se recorren todos los arcos y todos los nodos del grafo entregado.
